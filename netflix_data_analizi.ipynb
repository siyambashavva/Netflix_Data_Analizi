```python
# dizi ve film sayısının dağılımını verir.
sns.countplot(x="type", data=netflix)
plt.title("Tür Dağılımı")
plt.show()
```

```python


rating_counts = netflix["rating"].value_counts()

# rating sütunundaki benzersiz verilerin sayısı x ekseninde gösterilir.
sns.countplot(x=netflix["rating"], order=rating_counts.index)
plt.xticks(rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13, 13)
plt.title("Rating")
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.countplot(x = "rating", hue = "type", data = netflix)
plt.title("tür ve reyting oranları")
plt.show()
#dizi-film rating oranları aynı tabloda gösterilir.
```

```python
labels = ["Movie", "TV Show"]
size = netflix["type"].value_counts() # type sütunundaki benzersiz verilerin sayısını hesaplar.
colors = plt.cm.Wistia(np.linspace(0,1,2))
explode = [0, 0.1]
plt.rcParams["figure.figsize"] = (9,9)
plt.pie(size, labels = labels, colors = colors, explode = explode, shadow = True, startangle= 90)
plt.title("Tür Görünümü", fontsize= 25)
plt.legend()
plt.show()
```

```python
netflix["rating"].value_counts().plot.pie(autopct ="%1.1f%%", shadow= True, figsize= (10,8))
plt.show()
```
